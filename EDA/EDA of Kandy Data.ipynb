# Checking data with the Time

## Weather Data

### AT

```python
df_at_time = df_combined[['Period Start Time','AT']]
```

```python
df_at_time
```

```python
df_at_time.isnull().sum()
```

```python
sns.histplot(df_at_time)
```

```python
df_at_time['Period Start Time'].describe()
```

```python
df_at_time['Period Start Time'] = pd.to_datetime(df_at_time['Period Start Time'])
```

```python
df_at_time['Period Start Time']
```

```python
df_at_time['Date'] = df_at_time['Period Start Time'].dt.date
```

```python
df_at_time
```

```python
df_filtered1 = df_at_time[(df_at_time['Period Start Time'].dt.year >= 2019) &
                         (df_at_time['Period Start Time'].dt.year <= 2024)]
```

```python
daily_at = df_filtered1.groupby('Date')['AT'].mean().reset_index()
```

```python
plt.figure(figsize=(16, 8))
sns.lineplot(x='Date', y='AT', data=daily_at, linewidth=0.8)
plt.title('Monthly Mean AT from 2019 to 2024')
plt.xlabel('Date')
plt.ylabel('Mean AT')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_feb_2019 = df_at_time[
    (df_at_time['Period Start Time'].dt.year == 2019) &
    (df_at_time['Period Start Time'].dt.month == 2)
]
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_feb_2019, x='Date', y='AT', marker='o')
plt.title('Daily AT - Feburary 2019')
plt.xlabel('Date')
plt.ylabel('Ambient Temp')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_mar_2019 = df_at_time[
    (df_at_time['Period Start Time'].dt.year == 2019) &
    (df_at_time['Period Start Time'].dt.month == 3)
]
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_mar_2019, x='Date', y='AT', marker='o')
plt.title('Daily AT - March 2019')
plt.xlabel('Date')
plt.ylabel('Ambient Temp')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
